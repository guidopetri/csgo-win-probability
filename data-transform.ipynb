```python
from csgo.parser import FrameParser

parser = FrameParser('natus-vincere-vs-faze-m1-mirage.dem',
                     match_id='natus-vincere-vs-faze-m1-mirage')
```

```python
df = parser.parse()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df['areaId'].nunique()
```

```python
for c in ['X', 'Y', 'Z']:
    df[f'pos{c}'] = df[f'pos{c}'].astype(float)

df['areaId'] = df['areaId'].astype(int)
df['hp'] = df['hp'].astype(int)
```

```python
df.dtypes
```

```python
df.shape
```

```python
filtered_df = df[['side', 'name', 'posX', 'posY', 'posZ', 'tick', 'areaId']].copy()
#filtered_df.sort_values(by=['tick', 'side', 'name'], inplace=True)
filtered_df['pos'] = filtered_df[['posX', 'posY', 'posZ']].values.tolist()
filtered_df.head(10)
```

```python
import pandas as pd
merged = pd.merge(filtered_df, filtered_df, on='tick')
#merged.sort_values(by=['tick', 'side_x', 'name_x', 'side_y', 'name_y'], inplace=True)
merged['pos_diff'] = merged[['pos_x', 'pos_y']].values.tolist()
merged['area_diff'] = merged[['areaId_x', 'areaId_y']].values.tolist()

merged.head(7)
```

```python
from csgo.analytics.distance import point_distance, area_distance
```

```python
def graph_distance(x):
    return point_distance(x.values[0][0],
                          x.values[0][1],
                          map='de_mirage',
                          type='euclidean',
                          )

def area_dist_all(x, game_map):
    return area_distance(area_one=x.values[0][0],
                         area_two=x.values[0][1],
                         map=game_map,
                         )
```

```python
from functools import partial

area_dist = partial(area_dist_all, game_map='de_mirage')
```

```python
# tensor with dimensions:
# (tick, players_ct, players_t)
```

```python
import time
start_time = time.time()
sample = merged.pivot_table(index=['tick', 'name_x'],
                            columns='name_y',
                            values='area_diff',
                            aggfunc=area_dist)
print(time.time() - start_time)
```
